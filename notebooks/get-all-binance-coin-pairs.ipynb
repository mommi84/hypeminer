```python
import binance_bot

bot = binance_bot.BinanceBot('BTC', 'USDT') # any two will do

def get_binance_coins(base='USDT'):
    coins = []
    info = bot.get_exchange_info()
    for symbol in info['symbols']:
        s = symbol['symbol']
        if s.endswith(base):
            coins.append(s[:-len(base)])
    return coins

get_binance_coins()
```
